```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib.text import OffsetFrom
import numpy as np
from scipy.constants import golden as phi

# Read in data from csv
df = pd.read_csv("data/graphsheet_3.csv")
```

```python
# Sort dataframe by Ratio
df.sort_values(by='ratio',ascending=False, inplace=True)
```

```python
# Plot the ratios per department

# Initialize plot
fig, ax = plt.subplots(figsize=(16,10))
colorlist = [0.0,0.0,0.0,0.0]*9
scalar = [0.0]*9

# Plot Bars and Color per Ratio
for i in range (9):
    scalar[i] = float(df.iloc[i]['ratio'])/56.0
    colorlist[i] = colors.to_rgba((scalar[i], 1-scalar[i]*scalar[i], .05, scalar[i]*scalar[i]))
    ax = plt.bar(df.iloc[i]['dept_name'],df.iloc[i]['ratio'],1/phi,color=colorlist[i])

# Format the Plot Title and Labels
ax = plt.gca()
ax.set_title('Mentor : Replacement Ratio', fontsize=24)
ax.set_ylabel('Ratio of Mentors to Replacement Workers', fontsize=18)
ax.grid(axis="x", color="green", alpha=.3, linewidth=2, linestyle=":")
ax.grid(axis="y", color="black", alpha=.5, linewidth=.5)
plt.xticks(rotation=1/phi*100,fontsize=18)

# Plot the average value, which is 41
ax = plt.gca()
ax = plt.plot(([41]*9),linewidth=5)

# Annotate the average plot
plt.annotate('Average = 1:41', xy=(5.5, 42), xytext=(6.3, 50),
            arrowprops=dict(facecolor='black' , alpha=0.5), fontsize=23,bbox=dict(boxstyle="round", alpha=0.1))
fig.tight_layout()
fig.set_facecolor('lightgray') 

# Save Figure as SVG
plt.savefig("Resources/ratios.svg")

# Show and close plot
plt.show()
plt.close()
```
